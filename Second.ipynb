```python
l=[2,2,2]

nx.draw(nx.complete_graph(3), with_labels=1)

nx.is_graphical(l)
```

```python
P= nx.path_graph(5)

nx.draw(P, with_labels=1)
```

```python
# gives a list of cut vertex of the Graph
# cut vertex = the vertices without which graph becomes disconnected

list( nx.articulation_points(P) )
```

```python
G= nx.complete_graph(5)

H= nx.path_graph(6)

I= nx.disjoint_union(G, H)

nx.draw(I)
```

```python
nx.is_connected(I)
```

```python
S= nx.star_graph(5)

# NOTE: In a starGraph mid vertex is not counted from the #input vertices
nx.draw(S)

nx.is_connected(S)
```
